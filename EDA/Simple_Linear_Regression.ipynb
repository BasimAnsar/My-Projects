```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
hrs_studied = np.array([2, 4, 4, 6, 9, 7, 5, 8, 3, 7])
exam_scores = np.array([40, 55, 70, 85, 95, 60, 75, 80, 50, 65])
```

```python
hrs_studied = hrs_studied.reshape(-1,1)
```

```python
model=LinearRegression()
model.fit(hrs_studied,exam_scores)
```

```python
predicted_score=model.predict([[10]])
print(f"the predicted score is :{predicted_score[0]}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
predicted_score= model.predict(hrs_studied)
mae=mean_squared_error(exam_scores, predicted_score)
r2=r2_score(exam_scores, predicted_score)
```

```python
print("Mean Absolute Error:",mae)
print("Mean Squared Error:",mae)
print("R_squared score:",r2)
```

```python
plt.scatter(hrs_studied, exam_scores, label="Actual Scores")
plt.plot(hrs_studied, predicted_score, color="red", label="regression Line")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Scores")
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
hours_studied = np.array([2, 4, 3, 6, 9, 7, 5, 8, 3, 7])
attendance_percentage = np.array([75, 80, 78, 90, 95, 85, 82, 88, 76, 84])  # New feature
num_practice_tests = np.array([1, 3, 2, 4, 5, 3, 2, 4, 1, 3])              # New feature
exam_scores = np.array([40, 55, 70, 85, 95, 60, 75, 90, 50, 65])
```

```python
x=np.column_stack((hrs_studied, attendance_percentage, num_practice_tests))
y=exam_scores
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
y_pred=model.predict(x)
mse=mean_squared_error(y,y_pred)
r2=r2_score(y,y_pred)

print("Mean Squared Error:",mse)
print("R_squared score:",r2)
```

```python
student1=np.array([5,85,3])
student1_pred=model.predict([student1])
print(f"The predicted score for student1 is {student1_pred[0]}")
```
