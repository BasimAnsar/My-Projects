```python
#importing libraries
import pandas as pd
import seaborn as sns

#import data set
titanic_df=sns.load_dataset('titanic')
```

```python
print(titanic_df.head())
```

```python
mean_age=titanic_df['age'].mean()
print(f"mean age is : {mean_age: .2f} years")
```

```python
class_mode = titanic_df['class'].mode()[0]
print(f"Frequantly booked class is : {class_mode}class")
```

```python
fare_median = titanic_df['fare'].median()
print(f"median of the fare is: {fare_median: .2f} ")
```

```python
range_age = titanic_df['age'].max() - titanic_df['age'].min()
print(f"Range of the age is : {range_age: .2f} years")
```

```python
fare_deviation = titanic_df['fare'].std()
print(f"fare deviation is: {fare_deviation : .2f} dollars ")
```

```python
iqr_age = titanic_df['age'].quantile(0.75) - titanic_df['age'].quantile(0.25)
print(f"IQR of age is : {iqr_age: .2f} years")
```

```python
null_values= titanic_df['age'].isnull().sum()
print(f"null vlues: {null_values} Values")
```

```python
null_values= titanic_df[titanic_df['age'].isnull()]
print(f"The rows where 'age' is null are:\n{null_values}")
```

```python
import matplotlib.pyplot as plt
```

```python
def display_numarical_distribution(data,column,bins=20,kde=True):
  plt.figure(figsize=(8,6))
  sns.histplot(data[column].dropna(),bins=bins,kde=kde)
  plt.title(f"distrubition of{column}")
  plt.xlable(column)
  plt.ylable('frequency')
  plt.show()
```

```python
#print(display_numarical_distribution(titanic_df,column="age"))
display_numarical_distribution(titanic_df, column="age")
```
