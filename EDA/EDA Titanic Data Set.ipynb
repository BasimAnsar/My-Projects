```python
# import libraries
import pandas as pd
import seaborn as sns

#import the titanic dataset from seaborn library and put into data frame
titanic_df = sns.load_dataset('titanic')
```

```python
print(titanic_df.head())
```

```python
titanic_df['embarked'].unique()
```

```python
#titanic_df['age'].unique()
titanic_df['fare'].unique()
```

```python
mean_age = titanic_df['age'].mean()
print(f"Mean age is: {mean_age: .2f} years")
```

```python
median_fare = titanic_df['fare'].median()
print(f"Median fare is: {median_fare: .1f}")
```

```python
mode_embark= titanic_df['embarked'].mode()[0]
print(f"Mode embarked is: {mode_embark}")
```

```python
range_age = titanic_df['age'].max()-titanic_df['age'].min()
print(f" range of age is: {range_age} years")
```

```python
fare_deviation = titanic_df['fare'].std()
print(f"Fare deviation is: {fare_deviation: .2f}")
```

```python
iqr_age= titanic_df['age'].quantile(0.75) - titanic_df['age'].quantile(0.25)
print(f"IQR of age is: {iqr_age: .2f}")
```

```python
fare_variance = titanic_df['fare'].var()
print(f"Fare variance is: {fare_variance}")
```

```python
fare_skewness = titanic_df['fare'].skew()
print(f"Fare skewness is: {fare_skewness: .2f}")
```

```python
import matplotlib.pyplot as plt
sns.histplot(titanic_df['age'].dropna(), kde=True)
plt.title('Distribution of passangers Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='pclass',y= 'fare', data=titanic_df)
plt.title('fare pricess by the passanger class')
plt.xlabel('passenger class')
plt.ylabel('Fare')
plt.show()
```

```python
mean_age = titanic_df['age'].mean()
print(f"Mean age is: {mean_age: .4f} years")
```

```python
# Plot a histogram for survival based on age
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_df, x='age', hue='survived', multiple='stack', bins=30)
plt.title("Survival based on Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(["Did not survive", "Survived"])
plt.show()
```

```python
# Compare the mean ages of survivors and non-survivors
mean_age_survived = titanic_df[titanic_df['survived'] == 1]['age'].mean()
mean_age_not_survived = titanic_df[titanic_df['survived'] == 0]['age'].mean()

print(f"Mean Age of Survivors: {mean_age_survived:.2f} years")
print(f"Mean Age of Non-Survivors: {mean_age_not_survived:.2f} years")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_df = sns.load_dataset('titanic')
```

```python
titanic_df['age'].isnull().sum()
```

```python
titanic_df['age'].fillna(titanic_df['age'].median(),inplace=True)
```

```python
titanic_df['embarked'].isnull().sum()
```

```python
titanic_df['embarked'].fillna(titanic_df['embarked'].mode(),inplace=True)
```

```python
titanic_df['deck'].isnull().sum()
```

```python
titanic_df.drop('deck',axis=1,inplace=True)
```

```python
titanic_df['family_size'] = titanic_df['sibsp'] + titanic_df['parch'] + 1
```

```python
titanic_df['is_alone'] = 1
titanic_df['is_alone'].loc[titanic_df['family_size']>1]=0
```

```python
titanic_df.drop(['alive','who','adult_male', 'embark_town', 'class','sibsp','parch','alone'], axis=1,inplace=True)
```
